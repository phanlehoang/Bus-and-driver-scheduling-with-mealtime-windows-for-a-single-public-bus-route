```python
import pandas as pd
from datetime import datetime
from collections import namedtuple
from bus_driver_scheduling import BusDriverScheduling

# Đọc file Excel vào DataFrame
df = pd.read_excel('file_excel.xlsx')

# Chuyển đổi các cột thời gian sang định dạng datetime
df['time_start'] = pd.to_datetime(df['time_start'], format='%H:%M')
df['time_end'] = pd.to_datetime(df['time_end'], format='%H:%M')

# Chuyển DataFrame thành danh sách công việc
tasks = []
for row in df.itertuples(index=False):
    task = {
        'id': row.Id,
        'time_start': row.time_start,
        'time_end': row.time_end,
        'start_minute': row.start_minute,
        'end_minute': row.end_minute,
        'duration': row.duration
    }
    tasks.append(task)

# Khai báo danh sách tài xế
Driver = namedtuple('Driver', ['name', 'min_tasks', 'max_tasks'])
drivers = [
    Driver('Driver 1', 2, 4),
    Driver('Driver 2', 2, 4),
    Driver('Driver 3', 2, 3),
    Driver('Driver 4', 2, 3),
    Driver('Driver 5', 2, 3),
    Driver('Driver 6', 2, 3),
    Driver('Driver 7', 2, 3),
    Driver('Driver 8', 2, 3),
    Driver('Driver 9', 2, 3),
    Driver('Driver 10', 2, 3),
    Driver('Driver 11', 2, 3),
    Driver('Driver 12', 2, 3),
    Driver('Driver 13', 2, 3),
    Driver('Driver 14', 2, 3),
    Driver('Driver 15', 2, 3),
    Driver('Driver 16', 2, 3),
    Driver('Driver 17', 2, 3),
    Driver('Driver 18', 2, 3),
    Driver('Driver 19', 2, 3),
    Driver('Driver 20', 2, 3),
    Driver('Driver 21', 2, 3),
    Driver('Driver 22', 2, 3),
    Driver('Driver 23', 2, 3),
    Driver('Driver 24', 2, 3),
    Driver('Driver 25', 2, 3),
    Driver('Driver 26', 2, 3),
    Driver('Driver 27', 2, 3),
    Driver('Driver 28', 2, 3),
    Driver('Driver 29', 2, 3),
    Driver('Driver 30', 2, 3),
    Driver('Driver 31', 2, 3),
    Driver('Driver 32', 2, 3),
    Driver('Driver 33', 2, 3),
    Driver('Driver 34', 2, 3),
    Driver('Driver 35', 2, 3),
    Driver('Driver 36', 2, 3),
    Driver('Driver 37', 2, 3),
    Driver('Driver 38', 2, 3),
    Driver('Driver 39', 2, 3),
    Driver('Driver 40', 2, 3),
    Driver('Driver 41', 2, 3),
]

for info in tasks:
    info['time_start'] = datetime.strptime(str(info['time_start']), '%Y-%m-%d %H:%M:%S')
    info['time_end'] = datetime.strptime(str(info['time_end']), '%Y-%m-%d %H:%M:%S')

# Khởi tạo đối tượng BusDriverScheduling
scheduler = BusDriverScheduling(drivers, tasks)

# Tìm lời giải
scheduler.solve()

# Hiển thị lịch cho từng tài xế
for i, driver in enumerate(drivers):
    print(f"\nLịch của tài xế {i+1} ({driver.name}):")
    for j, task in enumerate(tasks):
        if scheduler.roster[j+1] == i:
            start_time_str = task['time_start'].strftime('%H:%M')
            end_time_str = task['time_end'].strftime('%H:%M')
            print(f"Công việc {task['id']} ({start_time_str} - {end_time_str})")
```
