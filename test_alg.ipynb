```python
from arc import Arc
from node import Node
# Create an node
node_1=Node(1, [Arc(cost = 1, travel_time = 10, start_point= 1, end_point =2), Arc(cost = 10, travel_time = 3, start_point= 1, end_point =3)])
node_2=Node(2, [ Arc(cost = 1, travel_time = 1, start_point= 2, end_point =4),
                Arc(cost = 2, travel_time = 3, start_point= 2, end_point =5)   ])
node_3=Node(3, [Arc(cost = 5, travel_time = 7, start_point= 3, end_point =4), Arc(cost = 1, travel_time = 2, start_point= 3, end_point =2),
                Arc(cost = 12, travel_time = 3, start_point= 3, end_point =5)])
node_4=Node(4, [Arc(cost = 1, travel_time = 7, start_point= 4, end_point =6), Arc(cost = 10, travel_time = 1, start_point= 4, end_point =5)])
node_5=Node(5, [Arc(cost = 2, travel_time = 2, start_point= 5, end_point =6)])
node_6=Node(6,[])


```

```python
#biến đống acrs thành từ điển
nodes = [node_1, node_2, node_3, node_4, node_5, node_6]

arcs_dict = {}
for node in nodes:
    for arc in node.arcs:
        arcs_dict[(arc.start_point, arc.end_point)] = arc
for keys in arcs_dict:
    print(keys, arcs_dict[keys])
```
